### FEATURES
The various features of the cleaned dataset are explained below:
1) Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines.
2) Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
3) Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities.
4) Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels.
5) Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6) Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
7) Destination City: City where the flight will land. It is a categorical feature having 6 unique cities.
8) Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy.
9) Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.
10)Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.
11) Price: Target variable stores information of the ticket price.

```python
## Feature Engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
## Drop Date Of Journey

df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.head(2)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
def standardize_duration(x):
    if 'h' in x and 'm' not in x:
        x = x + ' 0m'
    elif 'h' not in x and 'm' in x:
        x = '0h ' + x
    return x
df['Duration'] = df['Duration'].apply(standardize_duration)
```

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_hour']=df['Duration_hour'].astype(int)
df['Duration_min']=df['Duration_min'].astype(int)
df.drop('Duration',axis=1,inplace=True)
```

```python
df_new = df.drop(['Airline', 'Source', 'Destination'], axis=1)
df_new = pd.concat([df_new, encoded], axis=1)
df_new.to_csv("encoded_flight_data.csv", index=False)
df_new.head()
```

## Data Visualization

## Plot 1: Distribution of Flight Price

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_new['Price'], kde=True, bins=30, color='blue')
plt.title('Plot 1: Distribution of Flight Prices', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.savefig('1_price_distribution.png')
plt.show()
plt.close()
```

## Plot 2: Price vs. Total Stops

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Total_Stops', y='Price',data= df_new)
plt.title('Plot 2: Flight Price vs. Number of Stops', fontsize=16)
plt.xlabel('Total Stops', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.show()
```

## Plot 3: Price vs. Journey Month

```python
plt.figure(figsize=(12, 6))
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
             7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
monthly_avg = df_new.groupby('Month')['Price'].mean().reset_index().sort_values(by='Month')
monthly_avg['Month_Name'] = monthly_avg['Month'].map(month_map)
sns.lineplot(x='Month_Name', y='Price', data=monthly_avg, marker='o', sort=False)
plt.title('Plot 3: Average Flight Price by Month of Journey', fontsize=16)
plt.xlabel('Month of Journey', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.grid(True)
plt.show()
```

## Plot 4: Correlation Heatmap

```python
plt.figure(figsize=(14, 10))
corr_cols = ['Price', 'Total_Stops', 'Date', 'Month', 'Departure_hour', 'Arrival_hour', 'Duration_hour']
correlation_matrix = df_new[corr_cols].corr()
cmap = sns.diverging_palette(250, 10, s=90, l=30, n=9, as_cmap=True)
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap=cmap, 
    fmt=".2f", 
    linewidths=0.5, 
    cbar_kws={'shrink': 0.8}
)
plt.title('Plot 4: Correlation Matrix of Key Features', fontsize=16, fontweight='bold')
plt.show()
```

## Plot 5: Price vs. Duration

```python
plt.figure(figsize=(10, 6))
df_new['Total_Duration_Mins'] = df_new['Duration_hour'] * 60 + df_new['Duration_min']
sns.scatterplot(x='Total_Duration_Mins', y='Price', data=df_new, alpha=0.5, color='green')
plt.title('Plot 5: Flight Price vs. Total Duration of Flight', fontsize=16)
plt.xlabel('Total Duration (in Minutes)', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.tight_layout()
plt.savefig('5_price_vs_duration.png')
plt.show()
plt.close()
```
